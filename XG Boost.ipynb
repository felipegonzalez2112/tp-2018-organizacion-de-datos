```python
import pandas as pd
import matplotlib.pyplot as plt

import xgboost as xgb
from sklearn.model_selection import train_test_split
```

```python
#Cargo los archivos de entrenamiento
set_train = pd.read_csv("../set_a_entrenar-definitivo.csv")
set_predict = pd.read_csv("../set_a_predecir-definitivo.csv")
```

```python
set_train
```

```python
#Armamos arrays con los features que vamos a tener en cuenta
features=list(set_train.columns)
features
```

```python
#Eliminamos algunos features que consideramos irrelevantes
#features.remove('person')
features.remove('Unnamed: 0')
features.remove('person')
features.remove('label')
```

```python
#Lista de Features
features
```

```python
#Dividimos el set de entrenamientos para usar para la construccion del modelo y validarlo
train, validation=train_test_split(set_train,test_size=0.20, random_state=21)
print("TamaÃ±o set entrenamiento: ", len(train))
print("TamaÃ±o set de validacion: ", len(validation))
```

```python
#Empezamos a definir la variables x_train y_train 
x_train=train[features]
y_train=train['label']
x_validation = validation[features]
y_validation = validation['label']

```

```python
#Validamos nuestro modelo
model = xgb.XGBClassifier(base_score=0.5, booster='gbtree', colsample_bylevel=1,
       colsample_bytree=0.8, gamma=1, learning_rate=0.001,
       max_delta_step=0, max_depth=5, min_child_weight=5, missing=None,
       n_estimators=705, n_jobs=1, nthread=1, objective='binary:logistic',
       random_state=0, reg_alpha=0, reg_lambda=1, scale_pos_weight=1,
       seed=None, silent=True, subsample=0.8)
model.fit(x_train, y_train)
```

```python
#Hacemos un score del modelo

score = model.score(x_validation, y_validation)*100
print(score)
```

```python
test= set_predict[features]
test
```

```python
prediccion_final=model.predict_proba(test)
prediccion_final.size
```

```python
#Armamos el set de datos a submitear
submit = pd.read_csv('../trocafone_kaggle_test.csv', low_memory=False, index_col='person')
```

```python
#Como la prediccion final me da una dimension de 2 columnas. Me quedo con la segunda ya que es la probabilidad que el usuario realize una conversion
probabilidad_conversion=[]
for probabilidad in prediccion_final:
    probabilidad_conversion.append(probabilidad[1])
probabilidad_conversion
```

```python
#Agrego la columna label al set de datos de prediccion
submit['label'] = probabilidad_conversion
submit.head()
```

```python
#Analizo la proporcion de label
submit.label.mean()
```
